```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/Gaurang Vaghela/OneDrive/Desktop/Dataset/archive1/covid_vaccine_statewise.csv")
data
```

```python
data = data.drop(data[data["State"] == "India"].index)
data = data.drop(columns=['18-44 Years (Doses Administered)','45-60 Years (Doses Administered)','60+ Years (Doses Administered)',
                          '18-44 Years(Individuals Vaccinated)','45-60 Years(Individuals Vaccinated)','60+ Years(Individuals Vaccinated)',
                         'Male(Individuals Vaccinated)','Female(Individuals Vaccinated)','Transgender(Individuals Vaccinated)'])
data
```

```python
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
data.isnull().sum()
```

## Number of Persons State-Wise Vaccinated for First Dose

```python
first_dose = data.groupby("State")["First Dose Administered"].max().sort_values(ascending=False)
first_dose
```

```python
first_dose.plot(kind='barh', figsize=(12, 10), color='skyblue')
plt.title("State-wise First Dose Vaccination in India")
plt.xlabel("Number of People")
plt.ylabel("State")
plt.gca().invert_yaxis()
plt.show()
```

## Number of Persons State-Wise Vaccinated for Second Dose

```python
second_dose = data.groupby("State")["Second Dose Administered"].max().sort_values(ascending=False)
second_dose
```

```python
second_dose.plot(kind='barh', figsize=(12, 10), color='salmon')
plt.title("State-wise Second Dose Vaccination in India")
plt.xlabel("Number of People")
plt.ylabel("State")
plt.gca().invert_yaxis()
plt.show()
```

## Number of Males Vaccinated (Individuals)

```python
male_vaccinated = data.groupby("State")["Male (Doses Administered)"].max().sort_values(ascending=False)
male_vaccinated
```

```python
male_vaccinated.plot(kind='barh', figsize=(12, 10), color='blue')
plt.title("State-wise Male Individuals Vaccinated")
plt.xlabel("Number of Males")
plt.ylabel("State")
plt.gca().invert_yaxis()
plt.show()
```

## Number of Females Vaccinated (Individuals)

```python
male_vaccinated = data.groupby("State")["Female (Doses Administered)"].max().sort_values(ascending=False)
male_vaccinated
```

```python
male_vaccinated.plot(kind='barh', figsize=(12, 10), color='red')
plt.title("State-wise Female Individuals Vaccinated")
plt.xlabel("Number of Females")
plt.ylabel("State")
plt.gca().invert_yaxis()
plt.show()
```
